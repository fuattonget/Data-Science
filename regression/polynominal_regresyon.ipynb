```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python

# Importing the dataset
dataset = pd.read_csv('/Users/macbookpro/Documents/Data Science /regression/Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

```python
from sklearn.linear_model import LinearRegression

M = LinearRegression()

M.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_M = PolynomialFeatures(degree=4)

x_poly = poly_M.fit_transform(X)

model2 = LinearRegression()
model2.fit(x_poly,y)
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, M.predict(X) , color = 'blue')
plt.title('dogru yada yanıs (doğrusal regresyon)')
plt.xlabel('level')
plt.ylabel('ücret')

plt.show()
```

```python
plt.scatter(X,y, color = 'red')
plt.plot(X, model2.predict(x_poly) , color = 'blue')
plt.title('dogru yada yanıs (polinominal regresyon)')
plt.xlabel('level')
plt.ylabel('ücret')

plt.show()
```
